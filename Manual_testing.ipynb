```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from keras.preprocessing import image
```

```python
model = load_model('dog_modelv4.h5')
```

```python
test_image = image.load_img('Manual_Testing/Doberman.jpg', target_size=(256, 256))
test_image_array = image.img_to_array(test_image)
test_image_array = test_image_array / 255.0
test_image_array = np.expand_dims(test_image_array, axis=0) 
result = model.predict(test_image_array)
```

```python
Breed_Name = ['Bull Dog', 'Doberman', 'German Shepherd', 'Golden Retriever', 'Husky', 'Pariah Dog', 'Rottweilers']
print(f"Class Probabilities:", result[0])
predicted_Breed = np.argmax(result[0])
predicted_label = Breed_Name[predicted_Breed]
print("Predicted Class Label:", predicted_label)
for i, Probability in enumerate(result[0]):
    print(f"Class {Breed_Name[i]}: {Probability:.6f}")

print(f"Predicted Class Label: {predicted_label}")
```

```python
plt.imshow(test_image)
plt.title(f'Prediction: {predicted_label} ({result[0][predicted_Breed]*100:.6f}%)', fontsize=14)
plt.show()
```

```python
images = ['Manual_Testing/Bull_Dog.jpg','Manual_Testing/Doberman.jpg','Manual_Testing/German_Shepherd.jpg','Manual_Testing/Golden_Retriever.jpg','Manual_Testing/Husky.png','Manual_Testing/Pariah.jpg','Manual_Testing/Rottweiler.jpg']
```

```python
for i in images:
    test_image = image.load_img(i, target_size=(256, 256))
    test_image_array = image.img_to_array(test_image)
    test_image_array = test_image_array / 255.0
    test_image_array = np.expand_dims(test_image_array, axis=0) 
    result = model.predict(test_image_array)
    Breed_Name = ['Bull Dog', 'Doberman', 'German Shepherd', 'Golden Retriever', 'Husky', 'Pariah Dog', 'Rottweilers']
    predicted_Breed = np.argmax(result[0])
    predicted_label = Breed_Name[predicted_Breed]

    for i, Probability in enumerate(result[0]):
        print(f"Class {Breed_Name[i]}: {Probability:.6f}")
    plt.imshow(test_image)
    plt.title(f'Prediction: {predicted_label} ({result[0][predicted_Breed]*100:.6f}%)', fontsize=14)
    plt.show()
```
